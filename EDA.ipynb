```python
#plot distributions
k=1
plt.figure(figsize=(40, 40))
for col in df:
  if col=="Attrition":
    continue
  yes = df[df['Attrition'] == 'Yes'][col]
  no = df[df['Attrition'] == 'No'][col]
  plt.subplot(6, 6, k)
  plt.hist(yes, bins=25, alpha=0.5, label='yes', color='b')
  plt.hist(no, bins=25, alpha=0.5, label='no', color='r')
  plt.legend(loc='upper right')
  plt.title(col)
  k+=1
```

```python
df['Attrition'] = df['Attrition'].replace({'Yes': 1, 'No': 0})
df
```

```python
df.columns
```

```python
category = []
for col in df.columns:
    if df[col].dtype!='int64':
        category.append(col)
```

```python
print(category)
```

```python
# use get_dummies() method to one-hot encode the categorical data
df_1 = df
categorical_columns = category
for col in categorical_columns:
    col_type = pd.get_dummies(df[col], prefix=col)
    df_1 = pd.concat((df_1, col_type), axis=1).drop(col, axis=1)
```

```python
df_1.columns
```

```python
for col in df_1.columns:
    df_1[col] = df_1[col].replace({'True': 1, 'False': 0})
```

```python
df_1 = df_1.astype(int)
```
